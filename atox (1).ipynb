```python
#creating an array of o mixing ratio at different latitudes
dt=[alt_1,
alt_2,
alt_3,
alt_4,
alt_5,
alt_6,
alt_7,
alt_8,
alt_9,
alt_11,
alt_12,
alt_13,
alt_14,
alt_15,
alt_16,
alt_17,
alt_18,
alt_19,
alt_20,
alt_21,
alt_22,
alt_23,
alt_24,
alt_25,
alt_26,
alt_27,
alt_28,
alt_29,
alt_30,
alt_31,
alt_32,
alt_33,
alt_34,
alt_35,
alt_36,
alt_37,
alt_38,
alt_39,
alt_40,
alt_41,
alt_42,
alt_43,
alt_44,
alt_45,
alt_46,
alt_47,
alt_48,
alt_49]
degree = 5.5
myf=[]
for k in dt:

  coefficients = np.polyfit(height[:34], k[:34], degree)

  new_x = np.linspace(0, 350000, 1000)
  pre = np.polyval(coefficients, new_x)

  plt.ylim(0,2)
  plt.xlabel('altitude(m)')
  plt.ylabel('oxygen volume mixing ratio (mol/mol)')
# Plot the original data and the fitted curve
  plt.scatter(height, k, s=5)
  plt.plot(new_x, pre)
  myf.append(pre)
#avoiding negative o mixing ratio values from the model
plt.xlim(0,400000)
plt.ylim(0,2)
plt.show()

#ols fit and plotting the model

myf=np.array(myf)

#avoiding negative o mixing ratio values from the model
for k in range(len(myf)):
  for f in range(len(pre)):
    if myf[k][f]<0:
      myf[k,f]=0
```

```python

G=6.6* 10**-11 #universal grav const
area=200 #surface area of space craft
flu=[]
hfin=[]
#implementing the model in the trajectory of spacecraft
for f in range(len(halt)):
  for k in range(len(lat)):
    if lat[k]>= hlat[f]:
      for m in range(len(myf)):
        if lat[k]== lat[m]:
           o_col= myf[m]
      for j in range(len(new_x)):
           if new_x[j]>= halt[f]*1000:
             oxy= o_col[j]
             flux = area* np.sqrt(G*halt[f])*oxy #v= square root of G*h (orbital velocity of the spacecraft)
             break
      break
  if halt[f]<350:
      flu.append(flux)
      hfin.append(halt[f])
  else:
    flu.append(0)
    hfin.append(halt[f])


```

```python
fl=0
#integration of flux to get fluence
fluence=[]
for f in range(len(flu)):
  c= flu[f]+fl
  fl=c
  fluence.append(c)
#plotting fluence
plt.plot(time,fluence)
plt.xlabel('Time')
plt.ylabel('Fluence')
```
